```python
def changeStringToFloat(data_string):
    num = len(data_string)
    dim = len(data_string[0])
    data = []
    
    for i in range(num):
        if(data_string[i][2]!=''):
            temp = []
            for j in range(2,dim):
                temp.append(float(data_string[i][j].replace(",","")))
            data.append(temp)
    return data
```

```python
def judgeIncrease(data):
    num = len(data)
    for i in range(num-1):
        if(data[i]<data[i+1]):
            return 0
    return 1

print(judgeIncrease([3,2,3,4,5]))
print(judgeIncrease([5,4,3,2,1]))
```

```python
def abstractFeature(data):
    num = len(data)
    dim = len(data[0])
    feature = []
    
    
    EarningsPerShares = []
    for i in range(num):
        temp = []
        for j in range(5):
            temp.append(data[i][j])
        EarningsPerShares.append(temp)
    
    NetProfitTotalAssetRatio = []
    for i in range(num):
        temp = []
        for j in range(5,10):
            temp.append( data[i][j]/(data[i][j+10]+data[i][j+15])*10 )
        NetProfitTotalAssetRatio.append(temp)
        
    OperatingProfitTotalAssetRatio = []
    for i in range(num):
        temp = []
        for j in range(10,15):
            temp.append( data[i][j]/(data[i][j+5]+data[i][j+10]) )
        OperatingProfitTotalAssetRatio.append(temp)
    
    NetAssetTotalAssetRatio = []
    for i in range(num):
        temp = []
        for j in range(15,20):
            temp.append( data[i][j]/(data[i][j]+data[i][j+5]) )
        NetAssetTotalAssetRatio.append(temp)
    
    NetProfitIncrease = []
    for i in range(num):
        NetProfitIncrease.append( [judgeIncrease(data[i][10:15])] )
        
    OperatingProfitIncrease = []
    for i in range(num):
        OperatingProfitIncrease.append( [judgeIncrease(data[i][15:20])] )
        
    CompanySize = []
    for i in range(num):
        totalAsset = data[i][15]+data[i][20]
        if(totalAsset<=100000):
            tempSize = 0.0/5.0
        elif(totalAsset<=1000000 and totalAsset>100000):
            tempSize = 1.0/5.0
        elif(totalAsset<=10000000 and totalAsset>1000000):
            tempSize = 2.0/5.0
        elif(totalAsset<=100000000 and totalAsset>10000000):
            tempSize = 3.0/5.0
        else:
            tempSize = 4.0/5.0
        CompanySize.append([tempSize])
    feature = np.hstack((EarningsPerShares, NetProfitTotalAssetRatio, OperatingProfitTotalAssetRatio, NetAssetTotalAssetRatio, NetProfitIncrease,OperatingProfitIncrease, CompanySize))
    return feature
```

```python
feature = np.array(abstractFeature(data))
print(feature.shape)
print(feature)
```

```python
print(np.array(data).shape)

def generateLabel(data):
    num = len(data)
    dim = len(data[0])
    
    label = []
    label_combine = []
    
    for i in range(num):
        temp_label = 0
        max_price = np.max(data[i][25:30])
        min_price = np.min(data[i][25:30])
        price_2016 = data[i][30]
        
        point_1 = (max_price-min_price)/4+max_price
        point_2 = max_price
        
        if (price_2016>point_1):
            temp_label = 2
        elif (price_2016>point_2 and price_2016<point_1):
            temp_label = 1
        else:
            temp_label = 0
        label.append(temp_label)
        label_combine.append([temp_label])
        
    return label, label_combine
    
```

```python
import tensorflow as tf
train_data = np.array(feature)
train_label = np.array(label)
train_label_combine = np.array(label_combine)
train_combine = np.hstack((train_data,train_label_combine))

writer = csv.writer(open('./stock_data/proceeded_data.csv', 'w',newline=''))# you must add newline='' here, or there will be blank row
for i in train_combine:
    print(i)
    writer.writerow(i)
#writer.close()    
print(train_data.shape)
print(train_label.shape)
```

```python
train_data = np.float32(train_data)
#print(train_data)
train_label = np.int64(train_label)
hidden_layer_unit = 6

x_tf = tf.placeholder(tf.float32, shape=(None, train_data.shape[1]))
y_tf = tf.placeholder(tf.int64, shape=(None,))

#b1_tf = tf.Variable(tf.zeros([hidden_layer_unit]))
b1_tf = tf.Variable(np.zeros((hidden_layer_unit,)).astype('float32'))
W1_tf = tf.Variable(1e-2*np.random.rand(train_data.shape[1], hidden_layer_unit).astype('float32'))
#W1_tf = tf.Variable(tf.random_uniform([train_data.shape[1], hidden_layer_unit], -1.0, 1.0))
h = tf.nn.relu(tf.matmul(x_tf, W1_tf) + b1_tf)

b2_tf = tf.Variable(np.zeros((3,)).astype('float32'))
W2_tf = tf.Variable(1e-2*np.random.rand(hidden_layer_unit, 3).astype('float32'))
h_tf = tf.matmul(h, W2_tf) + b2_tf #    shape  (num, 3)

cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits= h_tf, labels=tf.one_hot(y_tf,3))
L2_loss = tf.nn.l2_loss(W1_tf) + tf.nn.l2_loss(W2_tf)
loss_tf = tf.reduce_mean(cross_entropy) + 0.005*L2_loss # prevent overfit

optimizer = tf.train.AdamOptimizer(1e-2)
train = optimizer.minimize(loss_tf)
init = tf.global_variables_initializer()

correct_num = tf.equal(y_tf,tf.argmax(h_tf,1)) # axis = 1, that's row
accuracy = tf.reduce_mean(tf.cast(correct_num,tf.float32))

sess = tf.Session()
sess.run(init)

for step in range(2000):
    sess.run(train, feed_dict={x_tf: train_data, y_tf: train_label})
    if step % 100 == 0:
        print(step,  sess.run(loss_tf, feed_dict={x_tf: train_data, y_tf: train_label}),sess.run(accuracy, feed_dict={x_tf: train_data, y_tf: train_label}) )
        #print(step,':',loss_tf)
sess.run(tf.argmax(h_tf,1), feed_dict={x_tf: train_data, y_tf: train_label})

saver = tf.train.Saver()
model_path = "./model_data/model.ckpt"
save_path = saver.save(sess, model_path)  
print ("Model saved in file: %s" % save_path) 
```
