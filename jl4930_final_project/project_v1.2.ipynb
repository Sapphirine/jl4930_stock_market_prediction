```python
import csv
import numpy as np
from function import *
import matplotlib.pyplot as plt
# import data from csv file
csv_reader = csv.reader(open('./stock_data/stock_02_train.csv'))

i=0
data_string = []
#data = np.array(data)
for row in csv_reader:
    #i=0
    if(i>=0):
        #print(row)
        data_string.append(row)
    i+=1
    
csv_reader = csv.reader(open('./stock_data/stock_02_train_sz.csv'))
for row in csv_reader:
    #i=0
    if(i>=0):
        #print(row)
        data_string.append(row)
    i+=1
    
row_number = i
print(np.shape(data_string))
#data_string[0]
```

```python
data = changeStringToFloat(data_string) #all data which is 
print(np.array(data).shape)
feature = np.array(abstractFeature(data))
print(feature.shape)
print(feature)
label,label_combine = generateLabel(data)
#label = np.reshape(label,(20,-1))
print(label)
print(np.array(label).shape)
```

```python
import tensorflow as tf
train_data = np.array(feature)
train_label = np.array(label)
train_label_combine = np.array(label_combine)
train_combine = np.hstack((train_data,train_label_combine))

writer = csv.writer(open('./stock_data/proceeded_train_data.csv', 'w',newline=''))# you must add newline='' here, or there will be blank row
for i in train_combine:
    print(i)
    writer.writerow(i)
#writer.close()    
print(train_data.shape)
print(train_label.shape)
```

```python
train_data = np.float32(train_data)   #train 
#print(train_data)
train_label = np.int64(train_label)
hidden_layer_unit = 20
label_num = 4

x_tf = tf.placeholder(tf.float32, shape=(None, train_data.shape[1]))
y_tf = tf.placeholder(tf.int64, shape=(None,))

#b1_tf = tf.Variable(tf.zeros([hidden_layer_unit]))
b1_tf = tf.Variable(np.zeros((hidden_layer_unit,)).astype('float32'),name = 'b1')
W1_tf = tf.Variable(1e-2*np.random.rand(train_data.shape[1], hidden_layer_unit).astype('float32'), name = 'W1')
#W1_tf = tf.Variable(tf.random_uniform([train_data.shape[1], hidden_layer_unit], -1.0, 1.0))
h = tf.nn.relu(tf.matmul(x_tf, W1_tf) + b1_tf)

b2_tf = tf.Variable(np.zeros((label_num,)).astype('float32'),name = 'b2')
W2_tf = tf.Variable(1e-2*np.random.rand(hidden_layer_unit, label_num).astype('float32'), name = 'W2')
h_tf = tf.matmul(h, W2_tf) + b2_tf #    shape  (num, 3)

cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits= h_tf, labels=tf.one_hot(y_tf, label_num))
L2_loss = tf.nn.l2_loss(W1_tf) + tf.nn.l2_loss(W2_tf)
loss_tf = tf.reduce_mean(cross_entropy) + 0.001*L2_loss # prevent overfit

optimizer = tf.train.AdamOptimizer(1e-2)
train = optimizer.minimize(loss_tf)
init = tf.global_variables_initializer()

correct_num = tf.equal(y_tf,tf.argmax(h_tf,1)) # axis = 1, that's row
accuracy = tf.reduce_mean(tf.cast(correct_num,tf.float32))

sess = tf.Session()
sess.run(init)

for step in range(5000):
    sess.run(train, feed_dict={x_tf: train_data, y_tf: train_label}) #the gradient will decrease
    if step % 200 == 0:
        print(step,  sess.run(loss_tf, feed_dict={x_tf: train_data, y_tf: train_label}),sess.run(accuracy, feed_dict={x_tf: train_data, y_tf: train_label}) )
        #print(step,':',loss_tf)
print(sess.run(tf.argmax(h_tf,1), feed_dict={x_tf: train_data, y_tf: train_label}))

saver = tf.train.Saver()
model_path = "./model_data/model.ckpt"
save_path = saver.save(sess, model_path)  
#print ("Model saved in file: %s" % save_path) 
```
